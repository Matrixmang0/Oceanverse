### Load Dataset: SOCR Height-Weight

This code reads a dataset from the file 'SOCR-HeightWeight.csv' into a pandas DataFrame named `data`. The dataset presumably contains information about height and weight.

This code removes the 'Index' column from the DataFrame `data` using the `drop` function along the specified axis (axis=1).

```python
data = data.drop('Index', axis = 1)
data
```

## Correlation Analysis

The code calculates the correlation matrix for the variables in the DataFrame `data` using the `corr()` method.

- **Correlation:**
  - Correlation measures the statistical association between two variables. It ranges from -1 to 1, where:
    - 1 indicates a perfect positive correlation (as one variable increases, the other also increases),
    - -1 indicates a perfect negative correlation (as one variable increases, the other decreases),
    - 0 indicates no linear correlation.

- **Correlation Formula:**
  - For two variables X and Y, the correlation (ρ) is calculated as:
    $ \rho_{XY} = \frac{cov(X, Y)}{\sigma_X \cdot \sigma_Y} $
    where $ cov(X, Y) $ is the covariance between X and Y, and $ \sigma_X $ and $ \sigma_Y $ are the standard deviations of X and Y.

- **Covariance Formula:**
  - Covariance (cov) between two variables X and Y is calculated as:
    $ cov(X, Y) = \frac{\sum{(X_i - \bar{X})(Y_i - \bar{Y})}}{n} $
    where $ \bar{X} $ and $ \bar{Y} $ are the means of X and Y, respectively, and n is the number of observations.

- **Types of Correlation:**
  - Positive Correlation: Both variables move in the same direction.
  - Negative Correlation: Variables move in opposite directions.
  - No Correlation: Variables show no linear relationship.

- **Significance of Slope:**
  - In a positive correlation, a positive slope indicates an increase in one variable corresponding to an increase in the other.
  - In a negative correlation, a negative slope indicates an increase in one variable corresponding to a decrease in the other.


```python
data.corr()
```

So, it can be found that height and weight are positively correlated with a correlation value of `0.502`. It means that as height of a person increases, their weight tends to increase as well.

## Feature and Label Extraction

The code extracts features and labels from the dataset `data`.

- **Features:**
  - The variable `features` is created from the 'Height(Inches)' column and reshaped to a one-dimensional array.

- **Labels:**
  - The variable `labels` is created from the 'Weight(Pounds)' column.


```python
features = np.array(data['Height(Inches)']).reshape(-1, 1)
labels = np.array(data['Weight(Pounds)'])
```

## Linear Regression Model

This code imports the `LinearRegression` class from scikit-learn and creates an instance named `lin_reg`. The model is then trained using the `fit` method with the extracted features and labels.


```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(features, labels)
```

## Linear Regression Prediction

This code predicts the values of the labels using the trained linear regression model (`lin_reg`) on the input features (`x_fit`). The predicted values are stored in the variable `y_fit`.  We will be using this values to plot the fitted line.


```python
x_fit = features
y_fit = lin_reg.predict(x_fit)
```

## Fitted Line Visualization

This code creates a scatter plot of height and weight data points, then overlays a fitted line generated by a linear regression model. The x-axis represents height (in inches), and the y-axis represents weight (in pounds).

```python
plt.scatter(data['Height(Inches)'], data['Weight(Pounds)'], s=2)
plt.plot(x_fit, y_fit, 'r-', label = 'Fitted line')
plt.xlabel("Height (Inches)")
plt.ylabel("Weight (Pounds)")
plt.legend();
```
